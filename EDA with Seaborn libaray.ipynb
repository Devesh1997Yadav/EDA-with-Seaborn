```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df =sns.load_dataset('tips')
```

```python
sns.set_style('whitegrid')
sns.distplot(df.total_bill,color = 'magenta')
```
